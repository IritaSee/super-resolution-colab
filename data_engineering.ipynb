```python
import os
import pydicom
from PIL import Image
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
main = '/content/patient_A'
savedir = '/content/drive/MyDrive/super-resolution-master/new_dataset/Y-90/patient_A/'
for a in os.listdir(main):
  for b in os.listdir(os.path.join(main,a)):
    dir = os.path.join(os.path.join(main,a,b))
    try:
      ds = pydicom.dcmread(dir)
      img = ds.pixel_array.astype(float)
      print("Converting to: ")
      print(os.path.join(savedir,a,b[:-4]+'.png'))
      scaled_image = (np.maximum(img, 0) / img.max()) * 255.0
      scaled_image = np.uint8(scaled_image)
      final_image = Image.fromarray(scaled_image)
      final_image.save(os.path.join(savedir,b[:-4]+'.png'))
    except:
      continue
```
